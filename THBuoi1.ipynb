```python
# PT B2
from math import *
def ptb2(a, b, c):
  delta = b**2-4*a*c
  if delta < 0 :
    print("Pt vo ngiem")
  elif(delta == 0) :
    print("Nghiem kep x1 = x2 = {0}".format(-b/(2*a)))
  else:
    print("Co hai nghiem x1 = {0}, x2 = {1}", (-b-sqrt(delta))/(2*a),(-b+sqrt(delta))/(2*a))
```

```python
a = int(input("Nhap a:"))
b = int(input("Nhap b:"))
c = int(input("Nhap c:"))
print(ptb2(a,b,c))
```

```python
def solve(n):
  tong = 0
  for i in range(1, n+1):
    if(i%2==0):
      tong -= i
    else:
      tong+=i

  return tong

print(solve(n = int(input("Nhap n:"))))
```

```python
dulieuLoad = load_iris()
dulieu_X = dulieuLoad.data[ : ]
dulieu_Y = dulieuLoad.target[ : ]
```

```python
X_Train, X_Test, Y_Train, Y_Test = train_test_split(dulieu_X, dulieu_Y, test_size=0.1, random_state=50, shuffle=True)
```

```python
# Load mo hinh
MoHinhDT = KNeighborsClassifier(n_neighbors=10)
MoHinhDT.fit(X_Train, Y_Train)
```

```python
Y_DuDoan = MoHinhDT.predict(X_Test)
print("Ket qua du doan :", Y_DuDoan)
print("Ket qua real:    ", Y_Test)
```

```python
KetQua_DoChinhXac = accuracy_score(Y_Test,Y_DuDoan)*100
print("Do chinh xac la : ", KetQua_DoChinhXac)
```

```python
X_Train, X_Test, Y_Train, Y_Test = train_test_split(dulieu_X, dulieu_Y, test_size=0.3, random_state=10, shuffle=True)
for e in range(2, 21):
  MoHinhDT = KNeighborsClassifier(n_neighbors=e)
  MoHinhDT.fit(X_Train, Y_Train)
  Y_DuDoan = MoHinhDT.predict(X_Test)
  KetQua_DoChinhXac = accuracy_score(Y_Test,Y_DuDoan)*100
  print("Do chinh xac la k = {0} = {1}",e, KetQua_DoChinhXac)

```
