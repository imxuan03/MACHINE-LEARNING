```python
import numpy as np
import matplotlib.pyplot as plt
# Dữ liệu mẫu
X = np.array([[1,2], [2,3], [4, 5], [5, 6]])
y = np.array([-1, -1, 1 , 1])

# Khởi tạo siêu phẳng ban đầu
w = np.array([0.5, 0.5])
b=0
# Hàm phân loại
def predict(X,w,b):
  return np.sign(np.dot(X,w)+b)

# Vẽ biểu đồ với các điểm dữ liệu và siêu phẳng ban đầu
plt.scatter(X[:,0], X[:,1], c = y, cmap=plt.cm.Paired, marker='o', s=100)
plt.axline((0, -b/w[1]), slope=-w[0]/w[1], linestyle='--', color='r', label='Siêu phẳng ban đầu')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Siêu phẳng SVM ban đầu')
plt.legend()
plt.grid(True)
plt.axis('equal')
plt.show()
```
